```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
titanic=pd.read_csv(r'Files\train.csv')
titanic.head()
```

```python
titanic.drop(labels=['PassengerId','Name','Ticket','Cabin','Embarked'],axis=1, inplace=True)
titanic.head()
```

```python
titanic.dtypes
```

```python
convertnumeric1={'male':1,'female':0}
titanic.Sex=titanic.Sex.map(convertnumeric1)
titanic.head()
```

```python
titanic.dtypes
```

```python
titanic.isnull().sum()
```

```python
titanic.shape
```

```python
titanic.fillna(value=titanic.Age.mean(), inplace=True)
titanic.isnull().sum()
```

```python
titanic.head(20)
```

```python
tree=DecisionTreeClassifier()
Survived=titanic['Survived']
Survived.head()
```

```python
#titanic.drop(['Survived'], axis=1, inplace=True)
titanic.head()
```

```python
#training the model by fit i.e 0 to 800
tree.fit(X=titanic[:800], y=Survived[:800])
tree.score(X=titanic[800:], y=Survived[800:])
```

```python
me=np.array([1,1,30,0,0,8])
me=me.reshape(1,-1)
tree.predict(me)
```
